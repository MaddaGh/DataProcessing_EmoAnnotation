### Competency questions

#### 1. Who is the person who appears to be expressing more emotions? What is their role?

Expected result: 

In the test data all agents are expressing 1 emotion, so it sould be<br>
**adelheid_rosen, guest, 1**<br>
**legien_kromkamp, host, 1**<br>
**anonymous, studio_audience, 1**<br>
**anonymous1, studio_audience, 1**<br>

```python
cq1 = """
    PREFIX : <http://www.semanticweb.org/ghiot/ontologies/2023/11/meaw/> 
    PREFIX meaw: <http://www.semanticweb.org/ghiot/ontologies/2023/11/meaw#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    
    SELECT ?agent ?role (COUNT(?emo) AS ?emo_count)
    WHERE {
        ?agent :hasRoleInConversation ?conv_role;
                :isSpeakerIn ?emo_rel.
        ?conv_role :hasRole ?role.
        ?emo_rel :expressesEmotion ?emo.
        ?emo a :BasicEmotion.
        
        FILTER(?emo != :neutrality)
    }

    GROUP BY ?agent
    ORDER BY DESC(?emo_count)
"""

results = modelet1.query(cq1)

for row in results:
    print(row)
    print("\n")
```

#### 2. Which utterances express a fine grained amotional state, beside a basic emotion? And what is it? 

Expected result: 

The utterance text and the emotional state retrieved.<br>
**Vrouw en wat ik nu zeg ..., nervousness**<br>
**Ja ik woon vragen je hoorde ..., aversion**<br>
**Dat zijn., argumentativeness**

```python
cq2 = """
    PREFIX : <http://www.semanticweb.org/ghiot/ontologies/2023/11/meaw/> 
    PREFIX meaw: <http://www.semanticweb.org/ghiot/ontologies/2023/11/meaw#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

    SELECT ?text ?emo
    WHERE {
        ?emo_rel :expressesEmotion ?emo;
                meaw:hasText ?text .
        ?emo rdf:type :EmotionState .
    }
    """
results = modelet1.query(cq2)

for row in results:
    print(row)
    print("\n")
```

#### 3. What are the conversational frames identified for the utterance starting at 612 seconds, and who is the speaker in the frame?

Expected result: 

A list of conversational frames the name of the speaker in  that utterance.<br>
**presenting_opinion**<br>
**anonymous1**<br>

```python
cq3 = """
    PREFIX : <http://www.semanticweb.org/ghiot/ontologies/2023/11/meaw/> 
    PREFIX meaw: <http://www.semanticweb.org/ghiot/ontologies/2023/11/meaw#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

    SELECT ?conv_frame ?speaker
    WHERE {
        ?emo_rel :hasStart "612";
                :hasFrameContext ?conv_frame ;
                meaw:hasText ?text .
        ?speaker :isSpeakerIn ?emo_rel .
    }
    """
results = modelet1.query(cq3)

for row in results:
    print(row)
    print("\n")
```

#### 4. What is the conversational frame where the host shows nervousness?

Expected result:

#### 6. What is the topic of the conversation and what is the talk show name? who is the broadcaster and what is the broadcasting date?

Expected result: 

A list of topics and the name of the talk show, the broadcaster name and the broadcasting date.<br>
**televisie, vrijheid van meningsuiting, welles nietes, ncrv, 1988-03-19**

```python
cq6 = """
    PREFIX : <http://www.semanticweb.org/ghiot/ontologies/2023/11/meaw/> 
    PREFIX meaw: <http://www.semanticweb.org/ghiot/ontologies/2023/11/meaw#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

    SELECT ?topic ?name ?broadcaster ?date
    WHERE {
        ?name rdf:type :TalkShowSituation;
                :isAbout ?topic ;
                :broadcastedBy ?broadcaster ;
                :hasDate ?date.
    }
    """
results = modelet1.query(cq6)

for row in results:
    print(row)
    print("\n")
```

#### 7. What are the gender of the angry speakers?

Expected result: 

The gender of the speaker and the number of the utterances in which they are angry<br>
**female, 1**

```python
cq7 = """
    PREFIX : <http://www.semanticweb.org/ghiot/ontologies/2023/11/meaw/> 
    PREFIX meaw: <http://www.semanticweb.org/ghiot/ontologies/2023/11/meaw#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

    SELECT ?gender (COUNT(?emo_rel) AS ?count) 
    WHERE {
        ?person :hasGender ?gender;
                :isSpeakerIn ?emo_rel.
        ?emo_rel :expressesEmotion :anger.
    }
    GROUP BY ?gender
    """
results = modelet1.query(cq7)

for row in results:
    print(row)
    print("\n")
```
