Expected result:

**55_anger, argumentativeness, conflict**<br>

```python
cq12 = """
    PREFIX : <https://raw.githubusercontent.com/MaddaGh/DataProcessing_EmoAnnotation/main/meaw.ttl#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    
    SELECT ?emo_rel ?emo ?vis_fr

    WHERE{
    ?emo_rel :hasVisualFraming ?vis_fr;
            :expressesEmotion ?emo.
    FILTER(?emo != :neutrality)
}

"""

results = modelet2.query(cq12)

for row in results:
    print(row)
    print("\n")
```

#### 13. What are the types of frames that compose utterances where Anger is shown?

Expected result: 

An utterance name, the camera angle, shot type, camera movement.<br>
**50_anger, horizontal_frontal, medium_close_up, stationary**<br>
**51_anger, horizontal_frontal, stationary, medium_close_up**<br>
**55_anger,horizontal_frontal, stationary, medium_close_up**<br>
**58_anger, horizontal_frontal, medium_long_shot, stationary**

```python
cq13 = """
    PREFIX : <https://raw.githubusercontent.com/MaddaGh/DataProcessing_EmoAnnotation/main/meaw.ttl#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    
    SELECT DISTINCT ?emo_rel ?angle ?shot ?mov

    WHERE{
        {
        ?emo_rel :expressesEmotion :anger;
                :framedBy ?frame.
        ?frame :hasAngle ?angle;
                :hasShotType ?shot;
                :hasMovement ?mov.
        }
        UNION
        {
        ?emo_rel :expressesEmotion ?emo;
                :framedBy ?frame.  
        ?emo :isIncludedIn :anger .
        ?frame :hasAngle ?angle;
                :hasShotType ?shot;
                :hasMovement ?mov.     
        }
        }

"""

results = modelet2.query(cq13)

for row in results:
    print(row)
    print("\n")
```
